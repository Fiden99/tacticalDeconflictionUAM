### functions to obtain various data

```python
def get_formatted_name(data_name,AP=False,NC=False, alpha=None, gapUnf=None):
    data_name = data_name.replace("Fixed", "")
    match = re.match(r"([a-zA-Z]+)(\d+)([a-zA-Z0-9-]*)", data_name)
    if match:
        part1 = match.group(1)
        part2 = match.group(3)
        part3 = ""
        part4 = ""
        if AP:
            part3 = "AP"
        elif NC:
            part3 = "NC"        
        if alpha is not None and gapUnf is not None:
            part4 = f"Alpha{alpha}Gap{gapUnf}"
        return f"{part1}{part2}{part3}{part4}"
    else:
        return data_name  # Caso in cui non ci sia un numero nel nome
```

```python
def obtainStatsSummary(filtered_merged, columnName, AP=False, NC=False):
    results = []
    grouped = filtered_merged.groupby(["alpha", "gapUnf"])
    for (alpha, gapUnf), group in grouped:
        finite_values = group[np.isfinite(group[columnName])][columnName]
        stats = {
            "Typology":         get_formatted_name(filtered_merged["instance"].iloc[0],AP,NC,alpha,gapUnf),
            "Mean":             finite_values.mean(),
            "STD":              finite_values.std(),
            "Min":              finite_values.min(),
            "Max":              finite_values.max(),
            "CountSolved":      finite_values.count(),
            "TotalCount":       group[columnName].count()
        }
        results.append(stats)
    return results
```

### Comparison between greater and smaller

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Set style
sns.set_style("whitegrid")

def plot_general_overview(datasets, names, metric='totalTime', title_suffix=''):
    """Plots distribution of a metric across multiple datasets"""
    plt.figure(figsize=(15, 6))
    
    data_to_plot = []
    labels = []
    
    for df, name in zip(datasets, names):
        # Filter valid data
        if metric in df.columns:
            valid_data = df[np.isfinite(df[metric])][metric]
            if len(valid_data) > 0:
                data_to_plot.append(valid_data)
                labels.append(name)
    
    # Boxplot
    plt.boxplot(data_to_plot, labels=labels, showmeans=True)
    plt.title(f'Distribution of {metric} {title_suffix}', fontsize=14)
    plt.ylabel(metric)
    plt.grid(True, alpha=0.3)
    plt.yscale('log') # Log scale often helps with time/opt differences
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# 1. General Overview of Time and Results
print("=== General Overview: Normal vs Smaller vs Greater ===")

# Prepare lists
dfs_dd = [pdDD, pdDD_Smaller, pdDD_Greater]
names_dd = ['DD Normal', 'DD Smaller', 'DD Greater']

dfs_ap = [pdAP, pdAP_Smaller, pdAP_Greater]
names_ap = ['AP Normal', 'AP Smaller', 'AP Greater']

dfs_nc = [pdNC, pdNC_Smaller, pdNC_Greater]
names_nc = ['NC Normal', 'NC Smaller', 'NC Greater']

# Plot Total Time
plot_general_overview(dfs_dd + dfs_ap + dfs_nc, names_dd + names_ap + names_nc, 'totalTime', '- All Datasets')

# Plot Optimization Value
plot_general_overview(dfs_dd + dfs_ap + dfs_nc, names_dd + names_ap + names_nc, 'opt', '- All Datasets')


# 2. Detailed Comparison: Smaller vs Greater
print("\n=== Detailed Comparison: Smaller vs Greater ===")

def compare_smaller_greater_detailed(df_smaller, df_greater, method_name):
    # Prepare data
    df_s = df_smaller.copy()
    df_s['Size'] = 'Smaller'
    df_g = df_greater.copy()
    df_g['Size'] = 'Greater'
    
    combined = pd.concat([df_s, df_g], ignore_index=True)
    
    # Filter
    combined = combined[np.isfinite(combined['opt']) & np.isfinite(combined['totalTime'])]
    
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))
    
    # Total Time
    sns.violinplot(data=combined, x='Size', y='totalTime', ax=axes[0], palette='muted', inner='quartile')
    axes[0].set_title(f'{method_name}: Total Time Distribution')
    axes[0].set_ylabel('Time (s)')
    
    # Opt
    sns.boxplot(data=combined, x='Size', y='opt', ax=axes[1], palette='muted')
    axes[1].set_title(f'{method_name}: Objective Value Distribution')
    axes[1].set_ylabel('Objective Value')
    axes[1].set_yscale('log')
    
    plt.suptitle(f'Comparison: {method_name}', fontsize=16)
    plt.tight_layout()
    plt.show()
    
    # Summary Table
    summary = combined.groupby('Size')[['totalTime', 'opt']].agg(['mean', 'median', 'std', 'count'])
    print(f"\nStats for {method_name}:")
    display(summary)

# Run comparisons
compare_smaller_greater_detailed(pdDD_Smaller, pdDD_Greater, "Delay & Drift (DD)")
compare_smaller_greater_detailed(pdAP_Smaller, pdAP_Greater, "High Priority (AP)")
compare_smaller_greater_detailed(pdNC_Smaller, pdNC_Greater, "Non-Collaborative Intruder(NC)")
```

### Comparison chaning gapUnf and alpha

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (14, 10)

# Function to analyze the effect of alpha and gapUnf parameters
def analyze_parameters_effect(df, dataset_name, metric='opt'):
    """Visualize how results change as alpha and gapUnf vary"""
    
    # Remove infinite values
    df_clean = df[np.isfinite(df[metric])].copy()
    
    # Create figure with 4 subplots
    fig = plt.figure(figsize=(16, 12))
    
    # 1. Heatmap - Average results for each combination
    ax1 = plt.subplot(2, 2, 1)
    pivot_mean = df_clean.groupby(['alpha', 'gapUnf'])[metric].mean().unstack()
    sns.heatmap(pivot_mean, annot=True, fmt='.2f', cmap='RdYlGn_r', ax=ax1, cbar_kws={'label': f'Average {metric}'})
    ax1.set_title(f'Heatmap: Average {metric} for alpha and gapUnf', fontsize=12, fontweight='bold')
    ax1.set_xlabel('gapUnf', fontsize=11)
    ax1.set_ylabel('alpha', fontsize=11)
    
    # 2. Line plot - Effect of alpha for each gapUnf
    ax2 = plt.subplot(2, 2, 2)
    for gapunf_val in sorted(df_clean['gapUnf'].unique()):
        subset = df_clean[df_clean['gapUnf'] == gapunf_val]
        means = subset.groupby('alpha')[metric].mean()
        ax2.plot(means.index, means.values, marker='o', label=f'gapUnf={gapunf_val}', linewidth=2, markersize=8)
    ax2.set_xlabel('alpha', fontsize=11)
    ax2.set_ylabel(f'Average {metric}', fontsize=11)
    ax2.set_title(f'Effect of alpha on {metric}', fontsize=12, fontweight='bold')
    ax2.legend(title='gapUnf', fontsize=9)
    ax2.grid(True, alpha=0.3)
    
    # 3. Line plot - Effect of gapUnf for each alpha
    ax3 = plt.subplot(2, 2, 3)
    for alpha_val in sorted(df_clean['alpha'].unique()):
        subset = df_clean[df_clean['alpha'] == alpha_val]
        means = subset.groupby('gapUnf')[metric].mean()
        ax3.plot(means.index, means.values, marker='s', label=f'alpha={alpha_val}', linewidth=2, markersize=8)
    ax3.set_xlabel('gapUnf', fontsize=11)
    ax3.set_xticks(sorted(df_clean['gapUnf'].unique()))
    ax3.set_ylabel(f'Average {metric}', fontsize=11)
    ax3.set_title(f'Effect of gapUnf on {metric}', fontsize=12, fontweight='bold')
    ax3.legend(title='alpha', fontsize=9)
    ax3.grid(True, alpha=0.3)
    
    # 4. Boxplot - Distribution for parameter combinations
    ax4 = plt.subplot(2, 2, 4)
    df_clean['params'] = '=' + df_clean['alpha'].astype(str) + ', gap=' + df_clean['gapUnf'].astype(str)
    param_order = sorted(df_clean['params'].unique(), key=lambda x: (int(x.split('=')[1].split(',')[0]), int(x.split('gap=')[1])))
    sns.boxplot(data=df_clean, x='params', y=metric, order=param_order, ax=ax4, palette='Set2')
    ax4.set_xlabel('Parameter Combination', fontsize=11)
    ax4.set_ylabel(metric, fontsize=11)
    ax4.set_title(f'{metric} Distribution by Parameter Combination', fontsize=12, fontweight='bold')
    ax4.tick_params(axis='x', rotation=45, labelsize=8)
    
    plt.suptitle(f'{dataset_name} - Parameter Analysis: alpha and gapUnf', fontsize=14, fontweight='bold', y=0.995)
    plt.tight_layout()
    
    # Display the plot
    plt.show()
    
    # Print summary statistics
    print(f"\n{'='*80}")
    print(f"SUMMARY STATISTICS: {dataset_name} - {metric}")
    print(f"{'='*80}\n")
    
    # Table with statistics for each combination
    summary_table = df_clean.groupby(['alpha', 'gapUnf'])[metric].agg(['mean', 'std', 'min', 'max', 'count'])
    summary_table = summary_table.round(2)
    print("Mean, standard deviation, min, max, and count for each combination:")
    display(summary_table)
    
    # Best combination
    best_idx = summary_table['mean'].idxmin()
    print(f"\n Best combination (minimum {metric}): alpha={best_idx[0]}, gapUnf={best_idx[1]}")
    print(f"  Average value: {summary_table.loc[best_idx, 'mean']:.2f}  {summary_table.loc[best_idx, 'std']:.2f}")
    
    # Worst combination
    worst_idx = summary_table['mean'].idxmax()
    print(f"\n Worst combination (maximum {metric}): alpha={worst_idx[0]}, gapUnf={worst_idx[1]}")
    print(f"  Average value: {summary_table.loc[worst_idx, 'mean']:.2f}  {summary_table.loc[worst_idx, 'std']:.2f}")
    
    print(f"\n{'='*80}\n")
    
    return summary_table

# USAGE EXAMPLES
# Analyze the effect of alpha and gapUnf on optimal value for the complete dataset
print("Analysis for Complete Dataset - 3 Delays (DD)")
analyze_parameters_effect(pdDD, 'Complete_Dataset_DD', metric='opt')

# Analyze the effect on total time
print("\n\nTotal Time Analysis for Complete Dataset (DD)")
analyze_parameters_effect(pdDD, 'Complete_Dataset_DD', metric='totalTime')

# You can repeat for other datasets by changing the dataframe:
# analyze_parameters_effect(df_grid01, 'grid01_DD', metric='opt')
# analyze_parameters_effect(df_mpAP, 'metroplex_AP', metric='opt')
# analyze_parameters_effect(df_G_air01, 'airport01_Greater', metric='totalTime')
```
